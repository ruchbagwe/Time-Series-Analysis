```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize']= 15, 6
```

```python
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m')
data = pd.read_csv('Data1\\Data1\\section 5\\AirPassengers.csv', parse_dates=['Month'], index_col='Month', date_parser=dateparse)
```

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
fit1 = ExponentialSmoothing(data, seasonal_periods=6, trend='add').fit(use_boxcox=True)
```

```python
fit1.fittedvalues.plot(style='--', color='red')
fit1.forecast(12).plot(style='--', marker='o', color='black', legend=True)
```
